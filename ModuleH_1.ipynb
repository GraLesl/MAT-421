## ODE Problem Statement

One extremley common task in mathematical and engineering disciplens are differential equations (DE). These DEs are equations which relate a function output based on an input, and any number of derivatives of said function.

For our purposes, we will be investigating the simplest classification of these, Ordinary Differential Equations (ODE). Here the independent variable and all the derivative terms are with respect to a single variable or dimension.

Modeling DEs is notoriously difficult as analytic solutions are far harder to come across than traditional functions. As such, digital analysis and approximation approaches are an extremley benificial tool when trying to model their behavior. A general solution would satisfy the DE at any given input, and these are very difficult to find. Furthermore, for an ODE with an order of n (the number of unique derivatives present in the DE), a particular soltion can be found given n explicitlly known values of the solution.

## Reduction of Order

Often times, it is easiest to solve initial value problems numerically with first order diferential equations (functions with no derivatives). However, these approaches can still be applied to higher order functions through reducing the order of a DE. This splits one higher order function into many first order functions that are all related. Often, it is useful to denote the state for initial value problems:

$S(t) = \begin{bmatrix} f(t) \\ f^{(1)}(t) \\ f^{(2)}(t) \\  f^{(3)}(t) \\ ...\\ f^{(n-1)}(t) \end{bmatrix} $

Then the derivative of this state can be written as

$\frac{dS(t)}{dt} = \begin{bmatrix} f^{(1)}(t) \\ f^{(2)}(t) \\  f^{(3)}(t) \\ ...\\ f^{(n)}(t) \end{bmatrix} = \begin{bmatrix} S_2(t) \\ S_3(t) \\  S_4(t) \\ ...\\ F(t,S_1(t),S_2(t),...,S_{n-1}(t) \end{bmatrix}$

Here, the states are effectivley coupled together where each equation is a first order ODE. Thereby succesfully reducing the order of an initial DE.

## The Euler Method

One approach to computing the results of an IVP is the Euler Mehtod. This method supposes that you are given a formula to approximate t he derivative of the state at a given time and state value: $\frac{dS(t)}{dt} = F(t,S(t))$. Given an initial time of $t_0 = 0$ and some final time $t_f = Nh$, where N is some positive integer and h is a given spacing. The subsequent  state can be written as $S(t_{j+1}) = S(t_j) + hF(t_j,S(t_j))$. As has been done in previous modeling problems, these small linear steps then comprise an estimation of the true funciton. An example of this can be seen below.

```python
import numpy as np
import matplotlib.pyplot as plt


f = lambda t, s: np.exp(-t)
h = 0.1
t = np.arange(0, 1 + h, h)
s0 = -1

s = np.zeros(len(t))
s[0] = s0

for i in range(0,len(t) - 1):
  s[i + 1] = s[i] + h*f(t[i],s[i])

plt.figure(figsize = (12,8))
plt.plot(t,s,'bo--', label='Approximate')
plt.plot(t, -np.exp(-t),'r', label='Exact')
plt.title('Euler Method - h = 0.01')
plt.xlabel('t')
plt.ylabel('s')
plt.legend()
plt.show()

f = lambda t, s: np.exp(-t)
h = 0.01
t = np.arange(0, 1 + h, h)
s0 = -1

s = np.zeros(len(t))
s[0] = s0

for i in range(0,len(t) - 1):
  s[i + 1] = s[i] + h*f(t[i],s[i])

plt.figure(figsize = (12,8))
plt.plot(t,s,'b--', label='Approximate')
plt.plot(t, -np.exp(-t),'r', label='Exact')
plt.title('Euler Method - h = 0.01')
plt.xlabel('t')
plt.ylabel('s')
plt.legend()
plt.show()

```
